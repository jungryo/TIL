```python
def forecast(lat, lon):
    url = "https://api.openweathermap.org/data/2.5/onecall?\
lat={}&lon={}&appid={}".format(lat, lon, KEY)
    response = requests.get(url)
    data = response.json()
    return data["timezone"], data["hourly"][1]
```
